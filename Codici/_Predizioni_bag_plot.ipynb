```python
import math
def round_half_up(n, decimals=3):
    multiplier = 10**decimals
    return math.floor(n * multiplier + 0.5) / multiplier
```

```python
def calcola_cdf(osservazioni):
    sorted_osservazioni = np.sort(osservazioni)
    n = len(sorted_osservazioni)
    cdf = np.arange(1, n + 1) / n
    return sorted_osservazioni, cdf
import seaborn as sns
plt.figure(figsize=(1, 1))
sns.kdeplot(mean16, fill=True,bw_method=0.05,cumulative=True,label=" ")
sns.kdeplot(mean1, fill=True,bw_method=0.05,cumulative=True,label=" ")
legend = plt.legend()

# Calcola la CDF dai dati di esempio
dati_ordinati16, cdf_calcolata16 = calcola_cdf(mean16.values)
dati_ordinati1, cdf_calcolata1 = calcola_cdf(mean1.values)
dati_ordinati16_s, cdf_calcolata16_s = calcola_cdf(datapd16["y_predict1"].values)
dati_ordinati1_s, cdf_calcolata1_s = calcola_cdf(datapd1["y_predict1"].values)

# Plot della CDF
plt.figure(figsize=(7, 6))  # Imposta le dimensioni del grafico
l=1.5
bb=40
# plt.plot(dati_ordinati1, cdf_calcolata1, label='CDF1',color="blue",linewidth=l)
plt.xticks(np.linspace(0,1,6))
#plt.grid(which="major",linestyle="--",alpha=0.5,color="grey",zorder=-5)

plt.fill_between(dati_ordinati16, cdf_calcolata16,alpha=0.23)
plt.fill_between(dati_ordinati1, cdf_calcolata1,alpha=0.23)
plt.plot(dati_ordinati16, cdf_calcolata16, linewidth=1)
plt.plot(dati_ordinati1, cdf_calcolata1, linewidth=1)

plt.legend(handles=legend.legendHandles,labels=["Ensemble with time-shift","Ensemble with NO time-shift"],loc="center left")
plt.ylabel("Cumulative Density Function", labelpad=10,fontsize=13)
plt.ylim(0,1.1)
plt.title("CDF of ensemble models' predictions\non defined polarity test set")
plt.savefig("/home/silvia/Desktop/Immagini/SIF_cimento/Cumulative_Instance.png", dpi=200)
```
